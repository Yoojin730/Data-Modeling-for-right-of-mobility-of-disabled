```python
from bs4 import BeautifulSoup
import requests
import re
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager  # 크롬드라이버 자동업데이트
```

```python
from urllib.request import HTTPError, urlopen 
import requests
from urllib import request
```

```python
url_head = "https://www.sisul.or.kr" # Html 불러오기    

titles = []
links = []

for i in range(53):
    url_board = url_head + "/open_content/calltaxi/qna/qnaMsgList.do?pgno=" + str(i)

    
    res = request.urlopen(url_board)
    soup = BeautifulSoup(res, "html.parser") # Beautifulsoup 사용해 html 분석

    # selector = "#detail_con > div.generalboard > table > tbody > tr > td.left.title > a"
    selector = "#detail_con > div.generalboard > table > tbody > tr > td.left.title > a"

    for a in soup.select(selector):
        titles.append(a.text)
        links.append(url_head + a.attrs["href"])
        
```

```python
import pandas as pd

board_df = pd.DataFrame({"title": titles, "link": links}) 
board_df
```

```python
contents=[]

for x in urls: 
    wd.get(x)
    time.sleep(3) 

    article = wd.find_element_by_xpath('//*[@id="detail_con"]/div[2]/table[1]/tbody/tr[3]/td').text
    article = article.replace('\n', "")
    article = article.replace('\t', "")
    article = article.replace('\\', "")
    article = article.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
    
    contents.append(article)    
        
```
