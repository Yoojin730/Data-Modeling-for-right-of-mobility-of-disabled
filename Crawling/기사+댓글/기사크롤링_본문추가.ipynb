```python
titles=[]
presses=[]
datetimes=[]
articles=[]

for x in urls: 
    wd.get(x)
    time.sleep(2) 
    
    try:
        press = wd.find_element_by_xpath('//*[@id="main_content"]/div[1]/div[1]/a/img').get_attribute('title') 
        title = wd.find_element_by_id('articleTitle').text
        datetime = wd.find_element_by_class_name('t11').text
        article = wd.find_element_by_id('articleBodyContents').text
        article = article.replace("function _flash_removeCallback() {}","")
        article = article.replace('\n', "")
        article = article.replace('\t', "")
        article = article.replace('\\', "")
        article = article.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
    
        titles.append(title)
        presses.append(press)
        datetimes.append(datetime)
        articles.append(article)    
        
    except :        
        titles.append('')
        presses.append('')
        datetimes.append('')
        articles.append('')    
```

```python
print(len(datetimes))
print(len(presses))
print(len(titles))
print(len(articles))
```

```python
d = {'발행일': datetimes, '언론사': presses, '기사URL' : urls, '제목' : titles, '본문' : articles }
df_final = pd.DataFrame(data=d)
df_final
```

```python
df_final = df_final[df_final['본문'] != '']
df_final = df_final.reset_index(drop=True)
df_final
```

```python
for i in range(len(df_final)):
    df_final['발행일'][i] = df_final['발행일'][i].split('.')[0] + '-' + df_final['발행일'][i].split('.')[1] + '-' + df_final['발행일'][i].split('.')[2]
```

```python
df_final['발행일'] = pd.to_datetime(df_final['발행일'])
```
