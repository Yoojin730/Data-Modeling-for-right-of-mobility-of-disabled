# 구별 장애인 인구 수2021

```python
pop2021 = pd.read_csv('/content/drive/Shareddrives/2022-01_데이터마이닝이론및응용/0_텀프로젝트/최종제출데이터/인구수/서울시동별장애인인구수2021.csv', encoding='cp949')
```

```python
pop2021.head()
```

```python
pop21 = pop2021[pop2021['동']=='소계']
```

```python
pop21.head()
```

```python
pop21.rename(columns={'기간':'연도'},inplace=True)
```

```python
pop21.rename(columns={'자치구':'구'},inplace=True)
```

```python
pop21.rename(columns={'합계':'인구수'},inplace=True)
```

```python
pop21_f = pop21[['구','인구수','연도']]
```

```python
pop21_f.head()
```

```python
pop21_f.info()
```

```python
pop21_f['인구수'] = pop21_f.인구수.str.replace(',', '').astype('int')
```

```python
pop21_f.sort_values(by=pop21_f.columns[1],ascending=False,inplace=True)
```
